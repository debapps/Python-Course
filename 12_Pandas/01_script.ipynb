# Python Pandas

- **Pandas** is a popular Python data analysis tool.

- It provides easy to use and highly efficient data structures.

- These data structures deal with numeric or labeled data, stored in the form of tables.

## Data Structures in Pandas

Two fundamental data structures used in pandas are:

1. **Series:** A 1-D array.

2. **Data Frame:** A 2-D array or two or more Series joined together.

```python
# importing pandas and numpy modules.
import numpy as np
import pandas as pd
```

### Series in Pandas

- **Series** is a 1-D array, holding data values of a single variable, captured from multiple observations.

- Few examples are:

    - Height of each student, belonging to a Class 'C'.

    - Amount of daily rainfall received at Station 'X', in July 2017.

    - Total sales of a product 'P' in every quarter of 2016.

```python
# Creating Series from python list.
# Height data list.
height_data = [4.3, 3.6, 5.1, 5.7, 6.1, 3.5, 3.9, 4.6]
height_series = pd.Series(height_data, index=['s1', 's2', 's3', 's4', 's5', 's6', 's7', 's8'])
height_series
```

```python
# Creating Series from numpy ndarry.
# Daily Rainfall in Kolkata in July, 2025.
kolkata_july_rainfall = pd.Series(10 * np.random.rand(31), index=[f'day{x}' for x in range(1, 32)])
print(kolkata_july_rainfall.head())
print(kolkata_july_rainfall.tail())
```

```python
# Creating Series from Python Dictionary.
# Total Quarterly Sales of in a grocery stores.
sales_data = {
    "Q1": 9000, "Q2": 11000, "Q3": 7800, "Q4": 15000
}

quarter_sales = pd.Series(sales_data)
quarter_sales
```

### DataFrame in Pandas

- A **DataFrame** is 2-D shaped and contains data of different parameters, captured from multiple observations.

- Each observation is represented by a single row, and each parameter by a single column.

- Each column can hold different data type.

- You can think of it like a spreadsheet or SQL table, or a dict of Series objects. 

- Few examples are:

    - Height and Weight of all students, belonging to a Class 'C'.

    - Daily  Rainfall received and Average Temperature of a location 'X', in the year 2017.

*DataFrame can be created from any of the following inputs:*

- From a Dict of 1-D ndarrays, lists, tuples, dicts or series.

- From the list of dictionaries.

- 2-D numpy ndarray.

- A Series.

- Data from spreadsheet (CSV or EXCEL), JSON file, SQL tables, Web Tables etc.

```python
# Creating DataFrame from Dict.

youtube_data = {
    'Channel': ['Code With Harry', 'Jym with Me', 'Explorer Nikk', 'Foodies Club'],
    'Ratings': (3.5, 3.2, 4.5, 4.7),
    'Likes': np.random.randint(1000, 9999, 4)
}
youtube_df = pd.DataFrame(youtube_data)
youtube_df
```

```python
# Creating DataFrame from list of Dictionaries.
books = [
    {'Name': 'The Time Machine', 'Author': 'H. G. Wells', 'Price in $': 2.3},
    {'Name': 'Sapiens', 'Author': 'Yuval Noah Harari', 'Price in $': 5.5},
    {'Name': 'Origin', 'Author': 'Dan Brown', 'Price in $': 7.5},
    {'Name': 'The Imortals of Meluha', 'Author': 'Amish', 'Price in $': np.nan}
]

book_df = pd.DataFrame(books, index = list('abcd'))
book_df
```

```python
# Creating DataFrame from 2-D numpy ndarray.
weather_data = np.array([[3.42, 33.2], [2.11, 25.7]])
weather_df = pd.DataFrame(weather_data, index=['Mumbai', 'Bangalore'], 
                          columns=['Average Rainfall', 'Average Temp in C'])
weather_df
```

```python
# Creating DataFrame from Series.
quarter_sales = pd.Series({
    "Q1": 9000, "Q2": 11000, "Q3": 7800, "Q4": 15000
}, name='Sales')
quarter_sales_df = pd.DataFrame(quarter_sales)
quarter_sales_df
```

## Accessing Data

- **Data Access** refers to extracting data present in defined data structures.

- **Pandas** provide utilities like **loc** and **iloc** to get data from a Series, a DataFrame, or a Panel.

### Accessing a Single Value

Individual elements can be accessed by specifying either **index number** or **index value**, inside the square brackets.

```python
z = np.arange(10, 16)
s = pd.Series(z, index=list('abcdef'))
s
```

```python
s[2]
```

```python
s['d']
```

```python
# Accessing value using get() method.
s.get(2)
```

```python
s.get('d')
```

#### Accessing a Slice

A Series can be sliced in a way, very similar to slicing a python list.

```python
s[1:4]
```

```python
# Elements corresponding to start and end index values are included, when index values are used for slicing.
s['b': 'e']
```

### Accessing DataFrame 

- DataFrame can be accessed as numpy ndarray or using the index and column labels.

```python
# Creating DataFrame from list of Dictionaries.
books = [
    {'Name': 'The Time Machine', 'Author': 'H. G. Wells', 'Price in $': 2.3},
    {'Name': 'Sapiens', 'Author': 'Yuval Noah Harari', 'Price in $': 5.5},
    {'Name': 'Origin', 'Author': 'Dan Brown', 'Price in $': 7.5},
    {'Name': 'The Imortals of Meluha', 'Author': 'Amish', 'Price in $': np.nan},
    {'Name': 'Deep Work', 'Author': 'Cal Newport', 'Price in $': 8.25}
]

book_df = pd.DataFrame(books)
book_df
```

```python
book_df['Author'][2]
```

```python
book_df.loc[2:]
```

```python
book_df.loc[1:3][['Name', 'Price in $']]
```

## Knowing a Series

- It is possible to understand a Series better by using **describe** method.

- The method provides details like mean, std, etc. about a series.

```python
rng = np.random.default_rng(seed=100)
temp = pd.Series(28 + 10 * rng.standard_normal(100))
temp.describe()
```

## Knowing a DataFrame

Two methods can be used to know about the data, present in a data frame.

    1. info()
    2. describe()

```python
weather_data = pd.DataFrame({
    'location': [f'loc_{idx}' for idx in range(1, 101)],
    'temp': pd.Series(28 + 10 * rng.standard_normal(100)),
    'rain': pd.Series(100 + 50 * rng.standard_normal(100))
})
weather_data.info()
```

```python
weather_data.describe()
```

```python
# You can use include argument to white list data types that has to be included in the result.
weather_data.describe(include=['object'])
```
