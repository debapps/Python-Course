```python
idx = [ 'r1', 'r2', 'r3', 'row4', 'row5', 'row6', 'r7', 'r8', 'r9', 'row10']
data = rng.integers(1, 100, 20).reshape(10, 2)
df = pd.DataFrame(data, index=idx, columns=['A', 'B'])
df
```

```python
g = df.groupby(df.index.str.len())
g.value_counts()
```

```python
g.filter(lambda x: len(x) > 1)
```

```python
g.aggregate({'A':len, 'B':np.sum})
```
