# Indexing In Pandas

- **Indexing** refers to labeling data elements of a Series, a Data Frame.

- These labels can be utilized for selecting portion of data from any of the defined data structures.

- A single level index can be set to a data frame, by passing a list of values to either using **index attribute** or **index argument** of DataFrame function.


```python
import pandas as pd
import numpy as np
rng = np.random.default_rng(seed=100)
```

```python
df = pd.DataFrame(np.random.rand(5,2))
df.index = [ 'row_' + str(i) for i in range(1, 6) ]
df
```

#### Different choices for indexing

- **.loc** is primarily label based, but may also be used with a boolean array. **.loc** will raise **KeyError** when the items are not found.

- **.iloc** is primarily integer position based **(from 0 to length-1 of the axis)**, but may also be used with a boolean array. **.iloc** will raise **IndexError** if a requested indexer is out-of-bounds, except slice indexers which allow out-of-bounds indexing.

```python
df.loc[['row_1','row_3']]
```

```python
df.loc[['row_5'], 1]
```

```python
df.iloc[:4]
```

```python
df.iloc[2:5][1]
```

```python
rng = np.random.default_rng(seed=100)
student_ids = [f's{idx}' for idx in range(1, 11)]
student_df = pd.DataFrame({
    'S_ID': student_ids,
    'Height': 170.0 + 25.0 * rng.standard_normal(10),
    'Weight': 75.0 + 12.0 * rng.standard_normal(10)
})
student_df
```

#### Boolean Indexing

- The operators are: 
    - **|** for **or**
    - **&** for **and** 
    - **~** for **not**

- These must be grouped by using parentheses. **(df['A'] > 2) & (df['B'] < 3)**.

- Boolean Indexing is used to filter data in the DataFrame.

```python
student_df.loc[student_df['S_ID'].isin(['s1', 's5'])]
```

```python
student_df[student_df['S_ID'].isin(['s1', 's5'])]
```

```python
student_df.iloc[3:7]
```

```python
student_df[student_df['Weight'] < 75]
```

```python
student_df[student_df['Height'] > 180][['S_ID', 'Height']]
```

```python
# Get the students whose height > 180 and weight < 75.
student_df[(student_df['Height'] > 180) & (student_df['Weight'] < 75)]
```

```python
# Get the students whose Height is not greater than 170.
student_df[~(student_df['Height'] > 170)][['S_ID']]
```

#### Example: Data Filtering.

**Task 1**

- Create a pandas series named **heights_A** with values: **176.2, 158.4, 167.6, 156.2, and 161.4**. These values represent the heights of 5 students of class A.

- Lebel each student as **s1, s2, s3, s4, and s5**.

- Create another series named **weights_A** with values: **85.1, 90.2, 76.8, 80.4, and 78.9**.These values represent the weights of 5 students of class A.

- Lebel each student as **s1, s2, s3, s4, and s5**.

- Create a dataframe named **df_A**, which holds the heights and weights of 5 students **s1, s2, s3, s4, and s5**.

- Label columns as **Student_height** and **Student_weight** respectively.

- Print the dataframe **df_A**.

**Task 2**

- Fiter the rows of **df_A**, whose **Student_height > 160.0** and **Student_weight < 80.0**, and capture them in another dataframe **df_A_filter1**.

- Print the dataframe **df_A_filter1**.

**Task 3**

- Filter the rows from **df_A**, whose index values end with 5, and capture them in another dataframe **df_A_filter2**.

```python
# Task 1.
student_labels = [f's{idx}' for idx in range(1, 6)]
heights_A = pd.Series([176.2, 158.4, 167.6, 156.2, 161.4], index = student_labels)
weights_A = pd.Series([85.1, 90.2, 76.8, 80.4, 78.9], index = student_labels)
df_A = pd.DataFrame({
    'Student_height': heights_A,
    'Student_weight': weights_A
}, index = student_labels)

df_A
```

```python
# Task 2.
df_A_filter1 = df_A[(df_A['Student_height'] > 160.0) & (df_A['Student_weight'] < 80.0)]
df_A_filter1
```

```python
# Task 3.
df_A_filter2 = df_A.loc[df_A.index.str.endswith('5')]
df_A_filter2
```

## Hierarchical indexing (MultiIndex)

- The **MultiIndex** object is the hierarchical analogue of the standard **Index** object which typically stores the axis labels in pandas objects. 

- **MultiIndex** can be thought as an array of tuples where each tuple is unique.

- A **MultiIndex** can be created from :
    - A list of arrays using `MultiIndex.from_arrays()`.
    - An array of tuples using `MultiIndex.from_tuples()`.
    - A crossed set of iterables using `MultiIndex.from_product()`.
    - A DataFrame using `MultiIndex.from_frame()`.


```python
# MultiIndex created from array.
array_index = [
    ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
    ["one", "two", "one", "two", "one", "two", "one", "two"],
]

mul_indx = pd.MultiIndex.from_arrays(array_index, names=['First', 'Second'])
mul_indx
```

```python
data = pd.DataFrame(rng.standard_normal(8), index=mul_indx)
data
```

```python
# MultiIndex created from tuple.
tuple_index = tuple(zip(*array_index))

mul_indx = pd.MultiIndex.from_tuples(tuple_index, name = ['First', 'Second'])
mul_indx
```

```python
# MultiIndex created from set of iterables.
iterables = [['Kolkata', 'Delhi', 'Mumbai', 'Chennai'], ['Max', 'Min']]
mul_indx = pd.MultiIndex.from_product(iterables, names=['City', 'Rank'])

rainfall_data = pd.DataFrame(140 + 15 * rng.standard_normal(8), index=mul_indx, columns=['Rainfall'])
rainfall_data
```

```python
# MultiIndex created from dataframe.
index_df = pd.DataFrame({
    'City': ['Kolkata', 'Kolkata', 'Delhi', 'Delhi', 'Mumbai', 'Mumbai', 'Chennai', 'Chennai'],
    'Rank': ['Max', 'Min', 'Max', 'Min', 'Max', 'Min', 'Max', 'Min']
})
mul_indx = pd.MultiIndex.from_frame(index_df)
data = {
    'Rain': 140 + 15 * rng.standard_normal(8),
    'Temp': 45 + 5 * rng.standard_normal(8)
}

weather = pd.DataFrame(data, index=mul_indx)
weather
```

#### Accessing Data with MultiIndex

```python
weather.loc[('Kolkata')]
```

```python
weather.loc[('Kolkata', 'Max')]
```

```python
weather.loc['Kolkata', 'Max']
```

```python
weather.loc[('Kolkata', 'Min'), 'Temp']
```

```python
# Sorting the MultiIndex.
weather.sort_index(inplace=True)
weather
```

```python
weather.loc['Chennai': 'Kolkata']
```

```python
weather.loc[('Delhi', 'Min') : ('Kolkata', 'Max')]
```

## DatetimeIndex in Pandas

- Immutable ndarray-like of **datetime64** data. Represented internally as **int64**.

#### Generating DatetimeIndex.

- `pandas.date_range()`: Returns a fixed frequency DatetimeIndex based on following parameters:
    
    - **start** : Staring Date (str)
    - **end** : Ending Date (str)
    - **Period** : Number of periods to generate (int)
    - **freq** : Date Time frequencies. **B** - Business Day, **D** - Calandar Day, **W** - Weekly, **ME** - Month end, **QE** - Quarter End, **YE** - Year End, **MS** - Month Start, .. etc. (Link: https://pandas.pydata.org/docs/user_guide/timeseries.html#offset-aliases)
    - **tz**: Timezones. ‘Asia/Hong_Kong’. By default,

- `pandas.bdate_range`: Return a fixed frequency DatetimeIndex with business day as the default.

```python
pd.date_range(start='1/1/2018', end='1/08/2018')
```

```python
pd.date_range(
    start=pd.to_datetime("1/1/2018").tz_localize("Europe/Berlin"),
    end=pd.to_datetime("1/08/2018").tz_localize("Europe/Berlin"),
)
```

```python
pd.date_range(start='1/1/2018', periods=8)
```

```python
pd.date_range(end='1/1/2018', periods=8)
```

```python
pd.date_range(start='2018-04-24', end='2018-04-27', periods=3)
```

```python
pd.date_range(start='1/1/2018', periods=12, freq='ME')
```

```python
pd.date_range(start='1/1/2018', periods=5, freq='3ME')
```

```python
# Business Days.
pd.bdate_range(start='1/1/2018', end='1/08/2018')
```

```python
# Generate sequences of fixed-frequency dates and time spans.
dti = pd.date_range(start='10/18/2025', periods=3, freq='h')
dti
```

```python
# Manipulating and converting date times with timezone information.
dti = dti.tz_localize('UTC')
dti
```

```python
dti.tz_convert('Asia/Kolkata')
```

```python
from datetime import datetime
# Use DatetimeIndex as Index.
data = pd.Series(rng.integers(low=1, high=20, size=10), index=pd.date_range(start=datetime.now(), periods=10, freq='h'))
data
```
