# Data Cleaning using Pandas

- Data Cleaning refers to transformation of raw data to a useful structured form, which is used for further analysis.

- The process includes various steps like dealing with missing values.

- Pandas provide following methods to deal with missing and duplicate data.

    - `dropna()`

    - `isna()`

    - `notna()`

    - `replace()`

    - `fillna()`

    - `drop_duplicates()`

#### `isna()` and `notna()` Methods

- `isna()` detects missing values (**NaN** in numeric arrays, **None** or **NaN** in object arrays, **NaT** in datetimelike) in array-like objects.

- `notna()` detects non-missing values for an array-like object.

```python
book_df.isna()
```

```python
book_df.notna()
```

#### `dropna()` Method.

- This method remove missing values.

##### Parameters


**axis: {0 or ‘index’, 1 or ‘columns’}, default 0**

- Determine if rows or columns which contain missing values are removed.

    - 0, or ‘index’ : Drop rows which contain missing values.

    - 1, or ‘columns’ : Drop columns which contain missing value.

    - Only a single axis is allowed.

**how: {‘any’, ‘all’}, default ‘any’**

- Determine if row or column is removed from DataFrame, when we have at least one NA or all NA.

    - ‘any’ : If any NA values are present, drop that row or column.

    - ‘all’ : If all values are NA, drop that row or column.

**subset: column label or sequence of labels, optional**

- Labels along other axis to consider, e.g. if you are dropping rows these would be a list of columns to include.

**inplace: bool, default `False`**

- Whether to modify the DataFrame rather than creating a new one.

**ignore_index: bool, default `False`**

- If `True`, the resulting axis will be labeled 0, 1, …, n - 1.

```python
book_df
```

```python
book_df.dtypes
```

```python
book_df.dropna(axis=0, how='any')
```

```python
# Drop all rows those have missing page number.
book_df.dropna(subset=['Page Number'])
```

```python
# Getting the unique values in the field - 'Price in dollar'.
book_df['Price in dollar'].unique()
```

```python
# Replacing the custom missing values ('NA', 'Missing') with np.nan.
book_df.replace('NA', np.nan, inplace=True)
book_df.replace('Missing', np.nan, inplace=True)
book_df
```

```python
book_df.dtypes
```

```python
# Droping the missing values in-place.
book_df.dropna(subset=['Book Title', 'Book Author', 'Price in dollar'], inplace=True)
book_df
```

```python
# Getting mean, median price of books.
print(f'Mean Price: {book_df['Price in dollar']. mean()}\nMedian Price: {book_df['Price in dollar'].median()}')
```

```python
# Fill the missing values with zeros.
book_df.fillna(0)
```

```python
# Drop Duplicate values based on Book Title and Book Author. Keeping first by default.
book_df.drop_duplicates(subset=['Book Title', 'Book Author'])
```

```python
# Drop Duplicate values based on Book Title and Book Author. Keeping last value.
book_df.drop_duplicates(subset=['Book Title', 'Book Author'], keep='last', 
                        inplace=True, ignore_index=True)
```
