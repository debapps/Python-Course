# Bar Plots & Pie Plots

## Bar Plot

- **Bar Plot** is commonly used for comparing categories.

- It is also used to compare categories over a short period of time.

- **bar** and **barh** are used for plotting vertical and horizontal bar plots respectively.

**Syntax**

```
bar(x,height)
# 'x' : x coordinates of bars.
# 'height' : List of heights of each bar.
```

```
barh(y, width)
# 'y' : y coordinates of bars
# 'width' : List of widths.
```

#### Bar Plot Example

- The below example plots the average sales of a company, in first three quarters of 2017.

- The code also sets the ticks on X-Axis and labels them.

````python
import matplotlib.pyplot as plt
import numpy as np 
````

````python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot()

ax.set(title='Avg. Quarterly Sales',
      xlabel='Quarter', ylabel='Sales (in millions)')

quarters = [1, 2, 3]
sales_2017 = [25782, 35783, 36133]

ax.bar(quarters, sales_2017)
ax.set_xticks(quarters)
ax.set_xticklabels(['Q1-2017', 'Q2-2017', 'Q3-2017'])

plt.show()
````

### Common Parameters of 'bar'

- **color**: Sets the color of bars.

- **edgecolor**: Sets the color of the border line of bars.

- **width**: Sets the width of bars

- **align**: Aligns the bars w.r.t x-coordinates. default - `center`.

    - `center`: Center the base on the x positions.

    - `edge`: Align the left edges of the bars with the x positions.

    - To align the bars on the right edge pass a negative width and `align='edge'`.

- **label**: Sets label to a bar, appearing in legend.

- **linewidth**: Width of the bar edge(s). If 0, don't draw edges.

````python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot()

ax.set(title='Avg. Quarterly Sales',
      xlabel='Quarter', ylabel='Sales (in millions)')

quarters = [1, 2, 3]
sales_2017 = [25782, 35783, 36133]

# Changing Color as cyan, edgecolor as magenta, width as 0.4.
ax.bar(quarters, sales_2017, width=0.4, color = 'c', edgecolor = 'm')
ax.set_xticks(quarters)
ax.set_xticklabels(['Q1-2017', 'Q2-2017', 'Q3-2017'])

plt.show()
````

### Plotting Multiple Groups

- Vertical bar plots are used for comparing more than one category at a time.

- The example below compares a company sales, occurred in first three quarters of 2016 and 2017.

````python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot()

ax.set(title='Avg. Quarterly Sales',
      xlabel='Quarter', ylabel='Sales (in millions)')

quarters = [1, 2, 3]
x1_index = [0.8, 1.8, 2.8]
x2_index = [1.2, 2.2, 3.2]
sales_2016 = [28831, 30762, 32178]
sales_2017 = [25782, 35783, 36133]

ax.bar(x1_index, sales_2016, color='yellow', width=0.4, edgecolor='black', label='2016')
ax.bar(x2_index, sales_2017, color='red', width=0.4, edgecolor='black', label='2017')

ax.set_xticks(quarters, labels=['Q1', 'Q2', 'Q3'])
ax.legend()

plt.show()
````

### Horizontal Plots: Barplot Using 'barh'

- `barh` draws the bars horizontally as shown in above image.

- **height** parameter is used to adjust the height of each bar.

- Horizontal bar plots are used while comparing values of one category at a time.

````python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot()

ax.set(title='Avg. Quarterly Sales',
      xlabel='Sales (in millions)', ylabel='Quarter')

quarters = [1, 2, 3]
sales_2017 = [25782, 35783, 36133]

ax.barh(quarters, sales_2017, height=0.6, color='blue')
ax.set_yticks(quarters, labels=['Q1-2017', 'Q2-2017', 'Q3-2017'])

plt.show()
````

- Define another list **sepal_len** with values **[6.01, 6.94, 7.69]**. These values represent the mean sepal length of iris flowers belonging to three species.

- Draw a **bar plot**, such that the height of each vertical bar displays the sepal length of a species label it as Sepal Length.

- Set bar **width as 0.4**, **color** needs to be set as **'#a4a4d8ff', '#800080', and '#f1d9e1ff'** respectively, and **border color** of the bar as **red**.

````python
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot()

# Axes Configuration.
ax.set(title ='Mean Sepal Length of Iris Species', 
       xlabel = 'Species', ylabel = 'Sepal Length (cm)',
       xlim = (0, 3), ylim = (0, 9))

# Data.
species = ['setosa', 'versicolor', 'virginica']
index = [0.4, 1.4, 2.4]
sepal_len = [6.01, 6.94, 7.69]

# Plotting.
ax.bar(index, sepal_len, width = 0.4, color = ['#a4a4d8ff', '#800080', '#f1d9e1ff'], 
       edgecolor = 'red', label = species)

ax.set_xticks(index, labels=species)
ax.legend()

plt.show()
````

````python
# Figure and Axes.
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot(111)

# Data.
sepal_len = [6.01, 6.94, 7.59]
sepal_wd = [4.42, 3.77, 3.97]
petal_len = [2.46, 5.26, 6.55]
petal_wd = [1.24, 2.33, 3.03]
species = ['setosa', 'versicolor', 'virginica']

# Data Indexes.
species_index1 = [0.7, 1.7, 2.7]
species_index2 = [0.9, 1.9, 2.9]
species_index3 = [1.1, 2.1, 3.1]
species_index4 = [1.3, 2.3, 3.3]

# Plot Configurations.
ax.set_title('Mean Measurements of Iris Species')
ax.set_xlim((0.5, 3.5))
ax.set_ylim((0, 10))
ax.set_xticks([1.0, 2.0, 3.0])
ax.set_xticklabels(species)
ax.set_xlabel('Species')
ax.set_ylabel('Iris Measurements (cm)')

# Plotting Data.
ax.bar(species_index1, sepal_len, color = 'm', edgecolor = 'gray', width=0.2, label = 'Sepal Length')
ax.bar(species_index2, sepal_wd, color = 'y', edgecolor = 'gray', width=0.2, label = 'Sepal Width')
ax.bar(species_index3, petal_len, color = 'c', edgecolor = 'gray', width=0.2, label = 'Petal Length')
ax.bar(species_index4, petal_wd, color = 'orange', edgecolor = 'gray', width=0.2, label = 'Petal Width')

# Adding Legend.
ax.legend()

plt.show()
````

#### Example

- Create a figure of size **15 inches** in width, and **5 inches** in height. 

- Define a list **species**, with elements **['setosa', 'versicolor', 'virginica']**.

- Define a list **index**, with values **[0.1, 1.1, 2.1]**.

- Define another list **petal_len** with values **[2.67, 5.49, 6.37]**. These values represent the mean petal length of iris flowers belonging to three species.

- Draw a **horizontal bar plot**, such that the width of each bar display the petal length of a species and label it as **Petal Length**.

- Use **index** and **petal_len** as variables. Set bar height as **0.4**, color as **m**, and border color of the bar as **c**.

- Label Y-Axis as **Species**. Label X-Axis as **Petal Length (cm)**.

- Set Title as **Mean Petal Length of Iris Species**.

- Mark major ticks on Y-Axis at **0.10, 1.10 and 2.10**. Label the major ticks on Y-Axis as **setosa, versicolor and virginica** respectively.

- Add a legend.

````python
# Figure and Axes.
fig = plt.figure(figsize=(15, 5))
ax = fig.add_subplot()

# Data.
species = ['setosa', 'versicolor', 'virginica']
index = [0.1, 1.1, 2.1]
petal_len = [2.67, 5.49, 6.37]
y_ticks = [0.10, 1.10, 2.10]

# Plot Configuration.
ax.set(title='Mean Petal Length of Iris Species', xlabel='Species',
       ylabel='Petal Length (cm)', yticks=y_ticks, yticklabels=species)

# Plotting.
ax.barh(index, petal_len, height=0.4, color = 'm', edgecolor = 'c', label = 'Petal Length')

# add legend.
ax.legend()

plt.show()
````

## Pie Plot

- **Pie plot** is effective in showing the proportion of categories.

- It is best suited for comparing fewer categories.

- In general, **Pie Plot** is used to highlight proportion of one or a group of categories.

**Syntax**

```
pie(x)
# 'x' : sizes of portions, passed either as a fraction or a number.
```

### Common Parameters of 'pie'

- **colors**: Sets the colors of portions.

- **labels**: Sets the labels of portions.

- **startangle**: Sets the start angle at which portion drawing starts.

- **autopct**: Sets the percentage display format of an area, covering portions.

- **counterclock**: Specify fractions direction, clockwise or counterclockwise. Default is `True`.

- **shadow**: If bool, whether to draw a shadow beneath the pie. If dict, draw a shadow passing the properties in the dict. Default is `False`.

````python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot()
ax.set_title('Avg. Quarterly Sales')

sales_2017 = [25782, 35783, 36133]
quarters = ['Q1-2017', 'Q2-2017', 'Q3-2017']

ax.pie(sales_2017,colors=['r', 'c', 'g'], labels=quarters, 
       startangle=90, counterclock=False, autopct='%1.1f%%',
       shadow=True)

plt.show()
````
