# Matplotlib Styles

- **matplotlib.pyplot** comes with a lot of styles. Based on the chosen style, the display of figure changes.

- The following commands shows the available styles in **pyplot**.

````python
import matplotlib.pyplot as plt
````

````python
plt.style.available
````

## Using a Style

A specific style can be invoked with either of the two expressions shown below.

`plt.style.use('ggplot')` 
or 
`plt.style.context('ggplot')`

When using *style context* the python *Context Manager* (**with** keyword) is recommended to use.

````python
with plt.style.context('ggplot'):

    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    
    ax.set(title='Avg. Daily Temperature of Jan 2018',
      xlabel='Day', ylabel='Temperature (in deg)',
      xlim=(0, 30), ylim=(25, 35))
    
    days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
    temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
    
    ax.plot(days, temp, color='green', linestyle='--', linewidth=3)
    plt.show()
````

- Multiple styles can be used simultenuously using a list.

`with plt.style.context(['dark_background', 'seaborn-v0_8-poster']):`

````python
with plt.style.context(['dark_background', 'seaborn-v0_8-poster']):

    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)
    
    ax.set(title='Avg. Daily Temperature of Jan 2018',
      xlabel='Day', ylabel='Temperature (in deg)',
      xlim=(0, 30), ylim=(25, 35))
    
    days = [1, 5, 8, 12, 15, 19, 22, 26, 29]
    temp = [29.3, 30.1, 30.4, 31.5, 32.3, 32.6, 31.8, 32.4, 32.7]
    
    ax.plot(days, temp, color='green', linestyle='--', linewidth=3)
    plt.show()
````

## Creating a Custom Style

- A style sheet is a text file having extension **.mplstyle**.

- All custom style sheets are placed in a folder, named **stylelib**, present in the config directory of **matplotlib**.

- Use the below expression for knowing the Config folder.

```
import matplotlib
print(matplotlib.get_configdir())
```

- Now, create a file **mystyle.mplstyle** with the below-shown contents and save it in the folder **<matplotlib_configdir>/stylelib/**. The <matplotlib_configdir> you will get from the expression result `matplotlib.get_configdir()`.

```
axes.titlesize : 24
axes.labelsize : 20
lines.linewidth : 8
lines.markersize : 10
xtick.labelsize : 16
ytick.labelsize : 16
```

- Reload the **matplotlib** library with the following expression.

`matplotlib.style.reload_library()``

- The below code snippet used **mystyle** along with **dark_background**.

```
with plt.style.context(['dark_background', 'mystyle']):
   ....
   ....
```

## matplotlibrc file

- **matplotlib** uses all the settings specified in **matplotlibrc** file.

- These settings are known as **rc settings** or **rc parameters**.

- For customization, **rc settings** can be altered in the file or interactively.

- The location of active **matplotlibrc** file used by matplotlib can be found with below expression.

```
import matplotlib as mpl
mpl.matplotlib_fname()
```

- All **rc settings**, present in **matplotlibrc** file are stored in a dictionary named **matplotlib.rcParams**.

- Any settings can be changed by editing values of this dictionary.

- For example, if you want to change `linewidth` and `color`, the following expressions can be used.

```
import matplotlib as mpl
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.color'] = 'r'
```

**Task 2**

- Regenerate the barplot using the style **'seaborn-v0_8-colorblind'**.

**Task 3**

- Regenerate the barplot using the style **'grayscale'**.

````python
# Generate the barplot.
def iris_barplots():
    fig = plt.figure(figsize=(9, 7))
    ax = fig.add_subplot()

    # Data.
    sepal_len = [6.01, 6.94, 7.59]
    sepal_wd = [4.42, 3.77, 3.97]
    petal_len = [2.46, 5.26, 6.55]
    petal_wd = [1.24, 2.33, 3.03]
    species = ['setosa','versicolor', 'virginica']
    
    species_index1 = [0.8, 1.8, 2.8]
    species_index2 = [1.0, 2.0, 3.0]
    species_index3 = [1.2, 2.2, 3.2]
    species_index4 = [1.4, 2.4, 3.4]

    # Plot Configuration.
    ax.set(title='Mean Mesurements of Iris Species',
           xlabel = 'Species', ylabel = 'Iris Measurements (cm)',
           xlim = (0.5, 3.7), ylim = (0, 10),
           xticks = species_index2, xticklabels = species)
    
    # Plots.
    ax.bar(species_index1, sepal_len, width=0.2, label = 'Sepal Length')
    ax.bar(species_index2, sepal_wd, width=0.2, label = 'Sepal Width')
    ax.bar(species_index3, petal_len, width=0.2, label = 'Petal Length')
    ax.bar(species_index4, petal_wd, width=0.2, label = 'Petal Width')
    ax.legend()

    plt.show()

# Task 1.
with plt.style.context('ggplot'):
    iris_barplots()
````

````python
# Task 2.
with plt.style.context('seaborn-v0_8-colorblind'):
    iris_barplots()
````

````python
# Task 3.
with plt.style.context('grayscale'):
    iris_barplots()
````
