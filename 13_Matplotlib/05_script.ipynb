# Histogram & Box Plots

## Histogram

- **Histogram** is used to visualize the spread of data of a distribution.

- **hist** function is used to plot a histogram.

**Syntax**

```
hist(x)
# 'x' : list of values or list of list of values.
```

#### Example

- Create a numpy array of student's grade percentages of **1000** students in normal distribution of mean **60** and standard deviation of **10**.

- Plot the distribution of student grades. Set the title as **Distribution of Student's Grade Percentage**, label x-axis as **Grade Percentage** and y-axis as **Student Count**.

````python
import numpy as np 
import matplotlib.pyplot as plt
rng = np.random.default_rng(seed=100) 
````

````python
# Student grade percentage data.
grades = 60 + 10 * rng.standard_normal(1000)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot()

ax.set(title = 'Distribution of Student\'s Grade Percentage',
       xlabel = 'Grade Percentage', ylabel = 'Student Count')

ax.hist(grades)
plt.show()
````

### Common Parameters of 'hist'

- **color**: Sets the color of bars.

- **bins**: Sets the number of bins to be used. If bins is a *sequence*, it defines the bin edges.

- **density**: Default value - `False`. When `True`, draw and return a probability density.

- **cumulative**: Default value - `False`. When `True`, then a histogram is computed where each bin gives the counts in that bin plus all bins for smaller values. If cumulative is a number less than `0` (e.g., `-1`), the direction of accumulation is *reversed*.

- **normed**: Sets to True where bins display fraction and not the count.

````python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot()

ax.set(title = 'Distribution of Student\'s Grade Percentage',
       xlabel = 'Grade Percentage', ylabel = 'Proportion')

# Applying bins = 30, drawing probability density function.
ax.hist(grades, color='g', bins=30, density=True)
plt.show()
````

#### Cumulative Probability Density Function

`cumulative = True`

````python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot()

ax.set(title = 'Distribution of Student\'s Grade Percentage',
       xlabel = 'Grade Percentage', ylabel = 'Proportion')

# Applying bins = 30, drawing probability density function.
ax.hist(grades, color='y',bins = 40, density=True, cumulative=True)
plt.show()
````

#### Example

- Create a figure of size **9 inches** in width, **7 inches** in height.

- Create a numpy array **dist_arr** of **1000** values in *normal distribution*, with mean **35** and standard deviation **3.0**.

- Draw a histogram of **dist_arr** with **35 bins**.

- Label X-Axis as **dist_arr**, Y-Axis as **Bin Count**.

- Set the title as **Histogram of a Single Dataset**.

````python
# Figure, Axes.
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot()

# Data.
np.random.seed(100)
dist_arr = np.random.normal(35, 3.0, 1000)

# Plot Configuration.
ax.set(title='Histogram of a Single Dataset', xlabel='dist_arr', ylabel='Bin Count')

# Plot.
ax.hist(dist_arr, bins=35)

plt.show()
````

## Box Plots

- **Box plots** are also used to visualize the spread of data.

- **Box plots** are used to compare distributions.

- **Box plots** can also be used to detect outliers.

- **Box plots** denotes the **five number summary** of the data.

    - *Minimum*
    - *1st Quartile (25%)*
    - *Median*
    - *3rd Quartile (75%)*
    - *Maximum*

**Syntax**
```
boxplot(x)
# 'x' : list of values or list of list of values.
```

#### Example

- Show the Box Plot of Grade Percentages of **1000** students in a Science Class.
- Set the following values:
    - title: **Box Plot of Student's Percentage**
    - X Label: **Science Class**
    - Y Label: **Percentage**

````python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot()

ax.set(title = 'Box Plot of Student\'s Grade Percentage',
       xlabel = 'Science Class', ylabel = 'Percentage')

# Box Plot.
ax.boxplot(grades)
plt.show()
````

### Common Parameters of 'boxplot'

- **notch**: Sets to `True`, if notches need to be created around the median.

- **bootstrap**: Number set to indicate that notches around the median are bootstrapped. *Values between 1000 and 10000 are recommended.*

- **orientation**: If `horizontal`, plots the boxes horizontally. Otherwise, plots the boxes vertically. Default value - `vertical`.

- **widths**: The widths of the boxes. The default is 0.5.

- **tick_labels**: The tick labels of each boxplot.

- **patch_artist**: Default value - `False`. If `True`, boxes are drawns as patch artist. *Used to color the boxes*.

````python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot()

ax.set(title = 'Box Plot of Student\'s Grade Percentage',
       xlabel = 'Science Class', ylabel = 'Percentage')

# Box Plot.
ax.boxplot(grades, notch=True, bootstrap=1000, tick_labels=['A'])
plt.show()
````

### Plotting Multiple Boxplots.

- There are three classes: **Science, Arts and Commerce**. The marks percentages of those classes are in *normal distribution* as follows:

    - **Science**: Mean - *75*, std - *25*
    - **Arts**: Mean - *40*, std - *15*
    - **Commerce**: Mean - *50*, std - *10*
    - Total Student Count - *500*

- Plot the above data in three box plots with levels as **Science, Arts and Commerce**. Set the following values:

    - title: **Box Plot of Student's Percentage**
    - X Label: **Class**
    - Y Label: **Percentage**

````python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot()

ax.set(title = 'Box Plot of Student\'s Grade Percentage',
       xlabel = 'Class', ylabel = 'Percentage')

# Student Marks Data.
labels = ['Science', 'Arts', 'Commerce']
np.random.seed(1000)
grades = [
    np.random.normal(75, 25, 500),
    np.random.normal(40, 15, 500),
    np.random.normal(50, 10, 50),
]

# Box Plot.
ax.boxplot(grades, tick_labels=labels, notch=True, bootstrap=1000)
plt.show()
````

### Plotting Boxplots Horizontally

`orientation = 'horizontal'`

````python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot()

ax.set(title = 'Box Plot of Student\'s Grade Percentage',
       ylabel = 'Class', xlabel = 'Percentage')

# Student Marks Data.
labels = ['Science', 'Arts', 'Commerce']
np.random.seed(1000)
grades = [
    np.random.normal(75, 25, 500),
    np.random.normal(40, 15, 500),
    np.random.normal(50, 10, 50),
]

# Box Plot.
ax.boxplot(grades, tick_labels=labels, notch=True, bootstrap=1000, orientation='horizontal')
plt.show()
````

#### Example

- Create a figure of size **9 inches** in width, **7 inches** in height.

- Create an array **arr_1** with **2000** values in *normal distribution* with mean **35** and standard deviation **6.0**.

- Create an array **arr_2** with **2000** values in *normal distribution* with mean **25** and standard deviation **4.0**.

- Create an array **arr_3** with **2000** values in *normal distribution* with mean **45** and standard deviation **8.0**.

- Create an array **arr_4** with **2000** values in *normal distribution* with mean **55** and standard deviation **10.0**.

- Draw boxplots with all the above 4 arrays with **nothes** and label them with **arr_1, arr_2, arr_3 and arr_4**.

- Fill colors inside the boxes using following color codes repectively: ['#F4F754', '#450693', '#FF3F7F', '#064232']

- Label X-Axis as **Dataset**, Y-Axis as **Value**. Set the title as **Box Plot of Multiple Datasets**.

````python
fig = plt.figure(figsize=(9, 7))
ax = fig.add_subplot()

ax.set(title = 'Box Plot of Multiple Datasets',
       ylabel = 'Dataset', xlabel = 'Value')

# Data.
np.random.seed(1000)
arr_1 = np.random.normal(35, 6.0, 2000)
arr_2 = np.random.normal(25, 4.0, 2000)
arr_3 = np.random.normal(45, 8.0, 2000)
arr_4 = np.random.normal(55, 10.0, 2000)
labels = ['arr_1', 'arr_2', 'arr_3', 'arr_4']
colors = ['#F4F754', '#450693', '#FF3F7F', '#064232']


# Box Plot.
plot = ax.boxplot([arr_1, arr_2, arr_3, arr_4], tick_labels=labels, notch=True, patch_artist=True)

# Fill colors.
for box, color in zip(plot['boxes'], colors):
    box.set_facecolor(color)

plt.show()
````
