# More about Subplots

- Till now, you have seen how to create a single plot in a figure.

- In this topic, you will see how to create multiple plots in a single figure.

- **subplot** is one of the functions used to create subplots.

**Syntax**

```
subplot(nrows, ncols, index)
```

- In the virtual grid in a **Figure**, `nrows` is the number of rows. `ncols` is the number of columns.

- `index` is the position in a virtual grid with `nrows` and `ncols`.

- **index** number varies from 1 to `nrows * ncols`.

- **subplot** creates the **Axes** object at **index** position and returns it.

````python
import matplotlib.pyplot as plt
import numpy as np 
````

````python
# Example
fig = plt.figure(figsize=(10,8))
axes1 = plt.subplot(2, 2, 1, title='Plot1')
axes2 = plt.subplot(2, 2, 2, title='Plot2')
axes3 = plt.subplot(2, 2, 3, title='Plot3')
axes4 = plt.subplot(2, 2, 4, title='Plot4')
plt.show()
````

````python
# Example
fig = plt.figure(figsize=(10,8))
axes1 = plt.subplot(2, 2, (1, 2), title='Plot 1', xticks = [], yticks = [])
axes2 = plt.subplot(2, 2, 3, title='Plot 2', xticks = [], yticks = [])
axes3 = plt.subplot(2, 2, 4, title='Plot 3', xticks = [], yticks = [])
plt.show()
````

````python
# Example
fig = plt.figure(figsize=(10,8))
axes1 = plt.subplot(2, 2, 1, title='Plot 1', xticks = [], yticks = [])
axes2 = plt.subplot(2, 2, (2, 4), title='Plot 2', xticks = [], yticks = [])
axes3 = plt.subplot(2, 2, 3, title='Plot 3', xticks = [], yticks = [])
plt.show()
````

## Subplots Using 'GridSpec'

- **GridSpec** class of **matplotlib.gridspec** can also be used to create Subplots.

- Initially, a grid with given number of rows and columns is set up.

- Later while creating a subplot, the number of rows and columns of grid, spanned by the subplot are provided as inputs to **subplot** function.

````python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(10,8))
gd = gridspec.GridSpec(2,1, figure=fig)

axes1 = plt.subplot(gd[0],title='Plot 1', xticks = [], yticks = [])

gd1 = gridspec.GridSpecFromSubplotSpec(1, 2, subplot_spec=gd[1])
axes2 = plt.subplot(gd1[0],title='Plot 2', xticks = [], yticks = [])
axes3 = plt.subplot(gd1[1],title='Plot 3', xticks = [], yticks = [])

plt.show()
````

````python
# Complex Example
fig = plt.figure(figsize=(10,8))
gd = gridspec.GridSpec(3, 3)

axes1 = plt.subplot(gd[0,:],title='Plot 1', xticks = [], yticks = [])
axes2 = plt.subplot(gd[1,:-1],title='Plot 2', xticks = [], yticks = [])
axes3 = plt.subplot(gd[1:,2],title='Plot 3', xticks = [], yticks = [])
axes4 = plt.subplot(gd[2,:-1],title='Plot 4', xticks = [], yticks = [])

plt.show()
````

### Example

- Define a numpy array **x** with numbers staring from **0.0**, stop at **10.0** step by **0.01**.

- Define a numpy array **s1** with expression: **sin(3 * pi *x)**.

- Define a numpy array **s2** with expression: **sin(6 * pi * x)**.

- Create a figure of size **12 inches** in width, and **7 inches** in height.

- Create an axes, named **axes1** that is the **first** virtual grid created by **2 rows** and **1 column**. Set the title as **Sin(3 * pi * x)**.

- Plot the line of **x** and **s1** using **axes1**.

- Create an axes, named **axes2** that is the **second** virtual grid created by **2 rows** and **1 column**. Set the title as **Sin(6 * pi * x)**.

- Set the arguments: **sharex** and **sharey** as **axes1**.

- Plot the line of **x** and **s2** using **axes2**.

````python
# Data.
x = np.arange(0.0, 10.0, 0.01)
s1 = np.sin(3 * np.pi * x)
s2 = np.sin(6 * np.pi * x)

# Plot.
fig = plt.figure(figsize=(12, 7))

# Axes.
axes1 = fig.add_subplot(211, title = 'Sin(3 * pi * x)')
axes2 = fig.add_subplot(212, title = 'Sin(6 * pi * x)', sharex = axes1, sharey = axes1)

# Plots.
axes1.plot(x, s1)
axes2.plot(x, s2)

plt.show()
````

#### Example

- Set numpy random seed to **1500**.

- Define a numpy array **x** with **10** random values between [0, 1).

- Define another numpy array **y** with **10** random values between [0, 1).

- Define another numpy array **z** with expression: **sqrt(x ** 2 + y ** 2)**.

- Create a figure of size **9 inches** in width, and **7 inches** in height.

- Create an axes, named **axes1**, which points the **first** virtual grid of the **2 rows** and **2 columns**. Set the title as **Scatter plot with Diamond Markers**.

- Draw a scatter plot of **x** and **y** using **axes1**. Set the following arguments:
    - **s to 80**
    - **c to z**
    - **marker to d**

- Add ticks on X-Axis at **0.0, 0.5, 1.0, 1.5** and on Y-Axis at **-0.2, 0.2, 0.6 and 1.0** respectively. 

- Create an axes, named **axes2**, which points the **second** virtual grid of the **2 rows** and **2 columns**. Set the title as **Scatter plot with Circle Markers**.

- Draw a scatter plot of **x** and **y** using **axes2**. Set the following arguments:
    - **s to 80**
    - **c to z**
    - **marker to o**

- Add ticks on X-Axis at **0.0, 0.5, 1.0, 1.5** and on Y-Axis at **-0.2, 0.2, 0.6 and 1.0** respectively. 

- Create an axes, named **axes3**, which points the **third** virtual grid of the **2 rows** and **2 columns**. Set the title as **Scatter plot with Plus Markers**.

- Draw a scatter plot of **x** and **y** using **axes3**. Set the following arguments:
    - **s to 80**
    - **c to z**
    - **marker to +**

- Add ticks on X-Axis at **0.0, 0.5, 1.0, 1.5** and on Y-Axis at **-0.2, 0.2, 0.6 and 1.0** respectively. 

- Create an axes, named **axes4**, which points the **fourth** virtual grid of the **2 rows** and **2 columns**. Set the title as **Scatter plot with Upper Triangle Markers**.

- Draw a scatter plot of **x** and **y** using **axes4**. Set the following arguments:
    - **s to 80**
    - **c to z**
    - **marker to 6**

- Add ticks on X-Axis at **0.0, 0.5, 1.0, 1.5** and on Y-Axis at **-0.2, 0.2, 0.6 and 1.0** respectively. 

- Adjust the entire layout with expression: `plt.tight_layout()`.

````python
# Data.
np.random.seed(1500)
x = np.random.rand(10)
y = np.random.rand(10)
z = np.sqrt(x ** 2 + y ** 2)

xtick_val = [0.0, 0.5, 1.0, 1.5]
ytick_val = [-0.2, 0.2, 0.6, 1.0]

# Figure.
fig = plt.figure(figsize=(9, 7))

# Axes.
axes1 = fig.add_subplot(221, title = 'Scatter plot with Diamond Markers',
                        xticks = xtick_val, yticks = ytick_val, 
                        xlim = (0.0, 1.5), ylim = (-0.2, 1.0))
axes2 = fig.add_subplot(222, title = 'Scatter plot with Circle Markers',
                        xticks = xtick_val, yticks = ytick_val,
                        xlim = (0.0, 1.5), ylim = (-0.2, 1.0))
axes3 = fig.add_subplot(223, title = 'Scatter plot with Plus Markers',
                        xticks = xtick_val, yticks = ytick_val,
                        xlim = (0.0, 1.5), ylim = (-0.2, 1.0))
axes4 = fig.add_subplot(224, title = 'Scatter plot with Upper Triangle Markers',
                        xticks = xtick_val, yticks = ytick_val,
                        xlim = (0.0, 1.5), ylim = (-0.2, 1.0))

# Plots.
axes1.scatter(x, y, c = z, s = [80], marker='d')
axes2.scatter(x, y, c = z, s = [80], marker='o')
axes3.scatter(x, y, c = z, s = [80], marker='+')
axes4.scatter(x, y, c = z, s = [80], marker=6)

plt.tight_layout()
plt.show()
````

#### Example

- Define a numpy array **x** with elements starting from **1**, stop at **301**, step by **3**.

- Define another numpy array **y1** with expression: **y1 = x**.

- Define another numpy array **y2** with expression: **y2 = x ** 2**.

- Define another numpy array **y3** with expression: **y3 = x ** 3**.

- Create a figure of size **9 inches** in width, and **7 inches** in height.

- Create a grid **gr** of **2 rows** and **2 columns**, using `GridSpec`. 

- Create an axes, named **axes1** spanning the **1st row** and **1st column** of grid **gr**. Set title as **y = x**.

- Draw a line plot of **x** and **y1** using **axes1**.

- Create an axes, named **axes2** spanning the **2nd row** and **1st column** of grid **gr**. Set title as **y = x ** 2**.

- Draw a line plot of **x** and **y2** using **axes2**.

- Create an axes, named **axes3** spanning all the rows of **2nd column** of grid **gr**. Set title as **y = x ** 3**.

- Draw a line plot of **x** and **y3** using **axes3**.

- Adjust the entire layout using `plt.tight_layout()` function.


````python
# Grid spec import.
import matplotlib.gridspec as gridspec

# Data.
x = np.arange(1, 301, 3)
y1 = x
y2 = x ** 2
y3 = x ** 3

# Figure.
fig = plt.figure(figsize=(9, 7))

# Creating Grid Layout.
gr = gridspec.GridSpec(2, 2)

# Creating Axes
axes1 = fig.add_subplot(gr[0, 0], title = 'y = x')
axes2 = fig.add_subplot(gr[1, 0], title = 'y = x ** 2')
axes3 = fig.add_subplot(gr[:, 1], title = 'y = x ** 3')

# Plotting.
axes1.plot(x, y1)
axes2.plot(x, y2)
axes3.plot(x, y3)

plt.tight_layout()
plt.show()
````

## Adding Text to the Subplot

- Text can be added to any plot using **text** function.

**Syntax**

```
text(x, y, s)
# 'x', 'y' represent x and y coordinates of a position.
# 's' is the text or string to be written
```

````python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)

ax.set(title='Writing Text',
      xlabel='X-Axis', ylabel='Y-Axis',
      xlim=(0, 5), ylim=(0, 9))

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

ax.scatter(x, y, c=['green'], s=[60], edgecolor='black')
for i in range(len(x)):
    str_temp = '({}, {})'.format(x[i] - 0.2, y[i] + 0.4)
    ax.text(x[i] - 0.4, y[i] + 0.4, str_temp, fontsize=16)
plt.show()
````
