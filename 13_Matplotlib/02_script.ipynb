# Anatomy of a Matplotlib Figure

- **Figure**: Whole area chosen for plotting.
- **Axes**: Area where data is plotted.
- **Axis**: Number-line like objects, which define graph limits.
- **Artist**: Every element on the figure is an artist.

## Figure

- **Figure** refers to the whole area or page on which everything is drawn.

- It includes Axes, Axis, and other Artist elements.

**Creating a Figure**

```
import matplotlib.pyplot as plt
fig = plt.figure()
```

````python
import matplotlib.pyplot as plt
````

````python
fig = plt.figure()
plt.show()
````

## Axes

- An **Axes** is the region of the figure, available for plotting data.

- An **Axes** object is associated with only one Figure.

- A **Figure** can contain one or more number of **Axes** elements.

- An **Axes** contains **two Axis objects** in case of 2D plots and **three Axis objects** in case of 3D plots.

- An Axes can be added to a figure using **add_subplot** methods.

`add_subplot(nrows, ncols, index)`

- When these argument values are less than 10, they all can be clubbed and passed as a single three-digit number. Hence, `add_subplot(1, 1, 1)` and `add_subplot(111)` are same.

````python
# figsize sets the size of the figure: width - 8 inches, height - 6 inches.
fig = plt.figure(figsize=(8, 6)) 
ax = fig.add_subplot(1, 1, 1)
plt.show()
````

````python
fig, axs = plt.subplots(ncols=2, nrows=2, figsize=(3.5, 2.5),
                        layout="constrained")
# for each Axes, add an artist, in this case a nice label in the middle...
for row in range(2):
    for col in range(2):
        axs[row, col].annotate(f'axs[{row}, {col}]', (0.5, 0.5),
                            transform=axs[row, col].transAxes,
                            ha='center', va='center', fontsize=18,
                            color='darkgrey')
````

#### Setting Title and Axis Labels

- **set** method can be used on created axes, **ax**, to set various parameters such as **xlabel, ylabel and title**.

````python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title='My First Plot',
      xlabel='X-Axis', ylabel='Y-Axis',
      xlim=(0, 5), ylim=(0,10))
plt.show()
````

- Setting an attribute can also be done with functions of the form **set_<parameter_name>**, as shown in below code.

````python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set_title('My First Plot')
ax.set_xlabel('X-Asis')
ax.set_ylabel('Y-Asis')
ax.set_xlim((0, 5))
ax.set_ylim((0, 10))
plt.show()
````

#### Plotting Data

- **plot** is one of the functions used for plotting data points.

- **plot** function is called on the created axes object, **ax**, as shown in below code snippet.

````python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title='My First Plot',
      xlabel='X-Axis', ylabel='Y-Axis',
      xlim=(0, 5), ylim=(0,10))
x = [1, 2, 3, 4]
y = [2, 5, 9, 3]
plt.plot(x, y)
plt.show()
````

#### Adding a Legend

- **legend** function is called on axes object ax to produce a legend.

- The legend uses the **label**, provided to a line drawn using **plot** as shown in below code.

````python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
ax.set(title='My First Plot',
      xlabel='X-Axis', ylabel='Y-Axis',
      xlim=(0, 5), ylim=(0,10))
x = [1, 2, 3, 4]
y = [2, 5, 6, 9]
plt.plot(x, y, label = 'Linear-Growth')
plt.legend()
plt.show()
````

#### Example

- Create a list **t** of values **7, 14, 21, 28, 35**.
- Create a list **d** of values so that **d = 6t**.
- Draw a line plot, by plotting **t** values on X-Axis and **d** values on Y-Axis. Label the line as **d = 6t**. Add a legend.
- Put X-Axis label as **time (seconds)** and Y-Axis label **distance (meters)**. 
- Set the title of the plot as **Time vs Distance Covered**.
- Limit data points on X-Axis from **0 to 40**, on Y-Axis from **0 to 220**. 
- Make the plot **12 inches in width**, and **6 inches in height**.

````python
# Initialize the plot.
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111)

# Data.
t = [7, 14, 21, 28, 35]
d = [item * 6 for item in t]

# Plot the data.
plt.plot(t, d, label = 'd = 6t')
plt.legend()

# Plot Configurations.
plt.title('Time vs Distance Covered')
plt.xlabel('time (seconds)')
plt.ylabel('distance (meters)')
plt.xlim((0, 40))
plt.ylim((0, 220))
plt.show()
````
