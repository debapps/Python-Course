## Central Tendencies

- Central Tendencies determines the center point or event of a dataset.

- There are three major central measurements: **mean**, **median** and **mode**.

### Mean

- **Mean** of dataset is calculated the sum of all values in that dataset, divided by the total number of value in that dataset.

- It denotes the average value.

- `mean` function of `numpy` is used to calculate mean.

```python
import numpy as np
from scipy import stats
```

```python
s1 = np.array([86, 75, 45, 47, 40])
print(np.mean(s1))
```

### Median

- **Median** is the middle value of the dataset when the data are arranged in ascending or descending order.

- If the dataset has an odd number of values, the median is the middle value. 

- If the dataset has an even number of values, the median is the average of the two middle values.

- `median` function of `numpy` is used to calculate median.


```python
s1 = np.array([86, 75, 45, 47, 40])
print(np.median(s1))
```

```python
s1 = np.array([86, 75, 45, 47, 40, 46])
print(np.median(s1))
```

### Mode

- **Mode** is the data point that comes the most frequently among a group of data members.

- `mode` function of `scipy.stats` module is used to calculate mode of the dataset.

- If there are more than a single value, then only the first one (in sorted order) is returned by mode function.

```python
s1 = np.array([86, 47, 45, 47, 40])
stats.mode(s1)
```

```python
s2 = np.array([86, 47, 45, 47, 40, 34, 45])
stats.mode(s2)
```

## Measures of Dispersion

- *Measures of Dispersion* provides the insights on how data is spread of a given dataset.

- Major measures of dispersion are: 
    - Range 
    - Percentile
    - Inter-quartile range (IQR)
    - Standard Deviation
    - Variance
    - Skewness
    - Kurtosis

### Range

- **Range** is the difference between the maximum and minimum values of the dataset.

- `ptp` function of the `numpy` module is used to calculate the range.

```python
s1 = np.array([86, 47, 45, 47, 41, 40, 36])
print(np.ptp(s1))
```

### Percentile

- **Percentile** refers to the value, below which the given percentage of the data points falls under the data set.

- For example: A student scored **85 percentile** in a test. This means you have performed better than **85%** of the candidates appeared in that test.

- Formula: `P(x) = (n/N) × 100` where:
    - P is the percentile,
    - n is Number of values below 'x',
    - N is Total count of population.

- `percentile` function of `numpy` can be used to compute a single or multiple percentiles.

#### Example:

What is the percentile of **80** in the following scores: **50, 55, 60, 40, 75, 75, 80, 90, 95, 100**?

#### Solution:

- Sorted data: **40, 50, 55, 60, 75, 75, 80, 90, 95, 100**.
- Number of falls under 80 (n) = 6
- Total number of count (N) = 10
- Percentile(x = 80) = (n/N) * 100 = 6/10 * 100 = 60.

**The percentile value of 80 in the given population is 60.**

#### Example:

What is the **15th** percentile for the given population of weights of persons **50, 55, 40, 60, 100, 95, 90, 60, 80, 75**?

#### Solution:

- Sorted data: 40, 50, 55, 60, 60, 75, 80, 90, 95, 100.

Given, Percentile (P) = 15

Total count of values (N) = 10

Percentile = (n/N) x 100

From the given formula we can find n by
n = (P x N)/100

= (15 x 10) / 100

= 150/100

= 1.5

1.5 can be rounded off to 2

And 2nd term in the sorted population is 50.
**15th percentile value is 50.**

```python
# Solve the problem using Python.
scores = np.array([50, 55, 40, 60, 100, 95, 90, 60, 80, 75])
np.percentile(scores, 15, method='lower')
```

```python
np.percentile(scores, 15, method='higher')
```

### Quartiles

- **Quartiles** divide a data set into four equal parts, each containing 25% of the data.

- Three **Quartiles** namely, **Q1**, **Q2**, and **Q3**, each part contains **25%** of total data.

- The same `percentile` function of `numpy` module is used to calculate quartiles of the dataset.

```python
scores = np.array([86, 47, 45, 47, 40, 97, 98, 75, 65, 83])
np.percentile(scores, [25, 50, 75], method='lower')
```

### Inter Quartile Range (IQR)

- **The interquartile range (IQR)** is a measure of spread of data, based on dividing a data set into quartiles.

- Unlike other measures of spread such as the range which can be heavily influenced by extreme values or outliers, the IQR isolates the central portion of the data.

- By calculating **the difference between the first and third quartiles (Q1 and Q3)** it gives us a clearer picture of how data points are distributed around the median.

`IQR = Q3 - Q1`

- `iqr` method from `scipy.stats` can be used for calculating it.


```python
scores = np.array([86, 47, 45, 47, 40, 97, 98, 75, 65, 83])
stats.iqr(scores, interpolation='lower')
```

### Variance and Standard Deviation

**Variance**

- **Variance** is defined as the average of squared differences, of each data point from dataset's mean.

`population variance = (x[i] - mean)**2 / N`

where N = Population size
    mean = Population mean

`sample variance = (x[i] - mean)**2 / (n - 1)`

where n = Sample size
    mean = Sample mean

**Standard Deviation**

- Standard Deviation is **square root of variance**.

- `var` and `std` functions of `numpy` can be used for computing variance and standard deviation respectively.

- By default, the functions assume that the dataset represents entire population.

- To represent a sample, derived from a population, `ddof` parameter is set to **1**.


```python
scores = np.array([86, 47, 45, 47, 40, 97, 98, 75, 65, 83])

# Population Variance and Standard Deviation.
print(f'Population Variance: {np.var(scores):.2f}')
print(f'Population Std: {np.std(scores):.2f}')

# Population Variance and Standard Deviation.
print(f'Sample Variance: {np.var(scores, ddof=1):.2f}')
print(f'Sample Std: {np.std(scores, ddof=1):.2f}')
```

### Skewness

- **skewness**   determines whether the majority of data points are present on one side of the distribution.

- A positive value represents **right skewed** distribution; a negative value represents **left skewed** one, zero represent **unskewed** distribution.

- The `skew` function of `scipy.stats` is used to calculate the skewness of the dataset.

```python
scores = np.array([86, 47, 45, 47, 40, 97, 98, 75, 65, 83])
stats.skew(scores)
```

```python
data = np.random.normal(loc = 10, scale = 0.6, size = 20)
stats.skew(data)
```

### Kurtosis

- **Kurtosis** indicates how much of data is concentrated around mean or shape of the probability distribution.

- It can be estimated using `kurtosis` function of `scipy.stats` module.

```python
scores = np.array([86, 47, 45, 47, 40, 97, 98, 75, 65, 83])
stats.kurtosis(scores)
```

#### Example

- Create a numpy array called `data` using normal ditribution of **mean - 150** and **standard deviation - 35.5** with size of **40** elements. (Use Random Seed as **1200**)

- Create a function `compute_descriptive_stats()`that accepts one parameter **data** and returns the following values in a list:

    - Mean
    - Median
    - Mode
    - 25th and 75th percentile or first and third quartiles **(Q1, Q2)** of the **data**.
    - Inter-quartile range value.
    - Skewness
    - Kurtosis

- Print those data in 2 decimal formats.

```python
import numpy as np
from scipy import stats

# Setting random data seed to 1200.
rng = np.random.default_rng(seed=1200)

# Generating Data Array.
data = rng.normal(loc=150, scale=35.5, size=40)

print('Data - \n', data)

# This function calculates the descriptive statistics of data.
def compute_descriptive_stats(data):
    mean = np.mean(data)
    median = np.median(data)
    mode = stats.mode(data).mode
    q1q3 = np.percentile(data, [25, 75], method='lower')
    iqr = stats.iqr(data)
    skew = stats.skew(data)
    kurtosis = stats.kurtosis(data)

    return [mean, median, mode, q1q3, iqr, skew, kurtosis]

print('\nDescriptive Statistics of Data:\n')

result = compute_descriptive_stats(data)
print(f'Mean: {result[0]:.2f}')
print(f'Median: {result[1]:.2f}')
print(f'Mode: {result[2]:.2f}')
q1q3 = result[3]
print(f'Q1: {q1q3[0]:.2f}')
print(f'Q3: {q1q3[1]:.2f}')
print(f'IQR: {result[4]:.2f}')
print(f'Skewness: {result[5]:.2f}')
print(f'Kurtosis: {result[6]:.2f}')
```

#### Describe the data.

- `describe()` function of `scipy.stats` is used to describe the statistics of dataset.

```python
des = stats.describe(data)

for key, val in des._asdict().items():
    print(f'{key} - {val}')
```
