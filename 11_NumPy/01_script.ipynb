# Introduction to Numpy

-   **N-dimensional array** is an object, capable of holding data elements of same type and of a fixed size in multiple dimensions.

```python
import numpy as np
```

### 1-D array creation.

```python
one_D_lst = [5, 6, 10, 12, 43, 65]
x = np.array(one_D_lst)
x
```

### 2-D array creation

```python
two_D_lst = [[3, 5, 65, -2], [9, 78, -98, 64]]
y = np.array(two_D_lst)
y
```

### Array Attributes:

-   **ndim** : Returns number of dimensions.

-   **shape** : Returns Shape in tuple.

-   **size** : Total number of elements.

-   **dtype** : Type of each element.

-   **itemsize** : Size of each element in Bytes.

-   **nbytes** : Total bytes consumed by all elements.

-   **flags** : Information about the memory layout of the array.

```python
print(f'Array Type: {type(x)}\nDimention: {x.ndim}\nShape: {x.shape}')
print(f'Size: {x.size}\nElement Type: {x.dtype}\nSize of Each Element(Bytes): {x.itemsize}')
print(f'Total bytes of all elements: {x.nbytes}\nMemory Info:\n{x.flags}')
```

```python
print(f'Array Type: {type(y)}\nDimention: {y.ndim}\nShape: {y.shape}')
print(f'Size: {y.size}\nElement Type: {y.dtype}\nSize of Each Element(Bytes): {y.itemsize}')
print(f'Total bytes of all elements: {y.nbytes}\nMemory Info:\n{y.flags}')
```

### Changing Datatype of Elements.

```python
y = np.array([[6, 9, 5],
              [10, 82, 34]],
             dtype=np.float64)
print(y)
print(y.dtype)
```

### ndarrays from Lists

```python
a = [[[4.1, 2.5], [1.1, 2.3], [9.1, 2.5]],
     [[8.6, 9.9],[3.6, 4.3], [6.6, 0.3]]]

x = np.array(a, dtype='float64')

print(type(x), x.ndim, x.shape)
```

### Array Creation Methods

```python
# Using zeros method.
x = np.zeros(shape=(2,3))
x
```

```python
# Using full method.
x = np.full(shape=(2, 4), fill_value=9.8)
x
```

```python
# Unsing zeros_like method.
arr = [[[2.3, 9.6], [23.6, 98.5], [3.4, 4.5]],
     [[3.14, 87.5], [7.4, 22.7], [8.6, 75.4]]]
print(arr)

x = np.zeros_like(a=arr)
print(x)
print(type(x), x.ndim, x.shape)
```

```python
# Using full_like method.
arr = [[3.42, 3.67], [9.87, 20.3]]
print(arr)

x = np.full_like(a=arr, fill_value=10.34)
print(x)
print(type(x), x.ndim, x.shape)
```

```python
# Using ones method.
x = np.ones(shape=(3,2))
print(x)
print(type(x), x.ndim, x.shape)
```

```python
# Using empty method.
x = np.empty(shape=(2,1))
x
```

```python
# np.eye(n, m) defines a 2D identity matrix. 
# The elements where i=j (row index and column index are equal) are 1 and the rest are 0.
x = np.eye(3)
x
```

### Random Numbers Generator

The method **default_rng()** is used to create a random number generator that is used to obtain samples from different distributions.

```python
# Generate one random float uniformly distributed over the range [0, 1).
rng = np.random.default_rng()
rng.random()
```

```python
# Generate random 1-D array where each element is in range [0, 1).
rng.random(size=5)
```

```python
# Generate random 2-D array where each element is in range [0, 1).
rand_arr1 = rng.random(size=(3, 4))
rand_arr1
```

```python
# Generate random 2-D array where each element is integer and in range 5-60.
rand_arr2 = rng.integers(5, 60, size=(4, 5))
rand_arr2
```

```python
# Generate an array of 10 numbers according to a unit Gaussian distribution.
rand_arr3 = rng.standard_normal(10)
rand_arr3
```

```python
# Generate an array of 10 numbers in normal distribution with mean 10 and standard deviation 2.
rand_arr4 = 10 + 2 * rng.standard_normal(10)
print(rand_arr4)
# Calculate Mean and Standard Deviation.
print(f'Mean: {rand_arr4.mean()}\nSTD: {rand_arr4.std()}')
```

```python
# Generate a binomial distribution of flipping a coin 100 times.
outcome = 1
probability = 0.5
total_trials = 100

samples = rng.binomial(n=outcome, p=probability, size=total_trials)
print(samples)

total_probability = sum(samples)/total_trials
print(total_probability)
```

##### Excercise 1

Write a function **arr** which can accept four parameters **a, x, y and z**, then perform the following operations:

- Define a ndarray **x1**, having first **a** even numbers ( Zero not included).
- Define a ndarray **x2**, having equally spaced **z** numbers between **x** and **y**.
- Print **x1** and **x2**.

```python
a, x, y, z = 41, 0, 30, 15

def arr(a, x, y, z):
    # Creating x1
    x1 = np.arange(2, a, 2)

    # Creating x2
    x2 = np.linspace(x, y, z)

    print(x1)
    print(x2)

arr(a, x, y, z)
```

##### Excercise 2

Write a function **array_3d** which can accept four parameters **n, x, y, z**, then perform the following:

- Simulate a random normal distribution for **n**, whose mean is **5** and standard deviation **2.5**. Capture the result in **x1**.
    - Find the product of the number of element in the array and the size of one array element in byte. Then store it in the variable **A**.
    **Note:** Set numpy seed value 100.

- Define a random 3-D array **x2** of shape **(x, y, z)** and of numbers **between 0 and 1**.

- Print the value of A and x2.

```python
n, x, y, z = 20, 3, 4, 2

def array_3d(n, x, y, z):
    mean, stddev = 5, 2.5

    # Random Generator.
    rng = np.random.default_rng(seed=100)

    # Normal Distribution.
    x1 = mean + stddev * rng.standard_normal(size=n)

    # Calculate total size.
    A = x1.size * x1.itemsize

    # Creating 3-D array.
    x2 = rng.random(size=(x, y, z))

    print(A)
    print(x2)

array_3d(n, x, y, z)
```

```python
n, x, y, z = 20, 3, 4, 2

def array_3d(n, x, y, z):    
    mean, stddev = 5, 2.5

    # Setting Seed to 100.
    np.random.seed(100)

    # Normal Distribution.
    x1 = mean + stddev * np.random.randn(n)

    # Calculate total size.
    A = x1.size * x1.itemsize

    # Creating 3-D array.
    x2 = np.random.random(size=(x, y, z))

    print(A)
    print(x2)

array_3d(n, x, y, z)
```
