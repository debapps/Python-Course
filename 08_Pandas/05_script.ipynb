# Data Aggregation

> **Aggregation:** compute a summary statistic (or statistics) for each group.

- By **“Data Aggregation”** we are referring to a process involving one or more of the following steps:

    - **Splitting** the data into groups based on some criteria.

    - **Applying** a function to each group independently.

    - **Combining** the results into a data structure.

```python
import pandas as pd
import numpy as np
rng = np.random.default_rng(seed=100)
```

#### Splitting an object into groups

```python
speeds = pd.DataFrame(
    [
        ("bird", "Falconiformes", 389.0),
        ("bird", "Psittaciformes", 24.0),
        ("mammal", "Carnivora", 80.2),
        ("mammal", "Primates", np.nan),
        ("mammal", "Carnivora", 58),
    ],
    index=["falcon", "parrot", "lion", "monkey", "leopard"],
    columns=("class", "order", "max_speed"),
)

speeds
```

```python
class_group = speeds.groupby(by = ['class'])
class_group.head()
```

```python
class_order_group = speeds.groupby(by = ['class', 'order'])
class_order_group.head()
```

```python
# MultiIndex Groups.
arrays = [
    ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
    ["doo", "doo", "bee", "bee", "bop", "bop", "bop", "bop"],
    ["one", "two", "one", "two", "one", "two", "one", "two"],
]
mul_indx = pd.MultiIndex.from_arrays(arrays, names=['First', 'Second', 'Third'])
data = pd.DataFrame(np.random.randn(8), index=mul_indx, columns=['Value'])
data
```

```python
data.groupby(by = ['Third']).sum()
```

```python
data.groupby(level=2).sum()
```

```python
data.groupby(by = ['First', 'Second']).sum()
```

#### Aggregation

An aggregation is a GroupBy operation that reduces the dimension of the grouping object

```python
animals = pd.DataFrame(
    {
        "kind": ["cat", "dog", "cat", "dog"],
        "height": [9.1, 6.0, 9.5, 34.0],
        "weight": [7.9, 7.5, 9.9, 198.0],
    }
)
animals
```

```python
animals.groupby(by = ['kind']).mean()
```

```python
animals.groupby("kind", as_index=False).sum()
```

### The `aggregate()` method

- Aggregate using one or more operations over the specified axis.

- It takes a function or list of functions and apply it on each groups of the group-by object.

- It takes positional and keyword arguments as well.

The main function signature is as follows:

`aggregate(func=None, *args, engine=None, engine_kwargs=None, **kwargs)`

```python
# Weather Data Generation.
cities = ['Kolkata', 'Delhi', 'Mumbai', 'Chennai']*3
np.random.shuffle(cities)

weather_data = {
    'City': cities,
    'Rainfall': pd.Series(120 + 30 * rng.standard_normal(12)),
    'Temperature': pd.Series(45 + 12 * rng.standard_normal(12))
}

weather_df = pd.DataFrame(weather_data)
weather_df
```

```python
weather_df.groupby(['City']).aggregate('mean')
```

```python
weather_df.groupby(['City']).aggregate(['mean', 'max', 'min', 'std'])
```

`apply()` method is used to apply the lambda function.

```python
weather_df.groupby(['City']).aggregate('mean').apply(lambda x: round(x, 2))
```
