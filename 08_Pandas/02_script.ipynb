# Input/Output with Pandas

Pandas provides support for reading/writing data from/to varieties of sources. 

#### CSV File

 - reader: **read_csv**
 - writer: **to_csv**

#### MS Excel

 - reader: **read_excel**
 - writer: **to_excel**

#### JSON File

 - reader: **read_json**
 - writer: **to_json**

#### HTML

 - reader: **read_html**
 - writer: **to_html**

#### XML

 - reader: **read_xml**
 - writer: **to_xml**

```python
import numpy as np 
import pandas as pd 
```

#### Reading and Writing CSV Files

```python
house_rent_df = pd.read_csv('./data/source/insurance.csv')
house_rent_df.head()
```

```python
# Get the csv file with header is None.
house_rent_df = pd.read_csv('./data/source/insurance.csv', header = None)
house_rent_df.head()
```

```python
# Skip rows
house_rent_df = pd.read_csv('./data/source/insurance.csv', skiprows = 2, header=None)
house_rent_df.head()
```

```python
# Random number generator.
rng = np.random.default_rng(seed = 100)

# Generating Data of students in the class of 10 students
students = [f's{idx}' for idx in range(1, 11)]
heights = pd.Series(170.0 + 25.0 * rng.standard_normal(10), index = students, name = 'Student_Height')
# print(f'Stident Heights:\n{heights}')

weights = pd.Series(75.0 + 12.0 * rng.standard_normal(10), index = students, name = 'Student_Weight')
# print(f'\nStudent Weight:\n{weights}')

# Genrating student dataframe.
std_df = pd.DataFrame({
    'Student_Height': heights,
    'Student_Weight': weights
}, index=students)

print(f'\nStudent Data:\n{std_df}')

# Writing student data into students_data1.csv
file = './data/target/students_data1.csv'
std_df.to_csv(file)
```

```python
# Writing student data into students_data2.csv with no index
file = './data/target/students_data2.csv'
std_df.to_csv(file, index=False)
```

#### Reading and Writing MS Excel File.

```python
# Reading the MS Excel Data.
src_file = './data/source/House_Rent.xlsx'
rental_data = pd.read_excel(src_file, sheet_name='House_Rent_Data')
print(f'All the Data Columns:\n {rental_data.columns}')

# Get only required columns.
house_rent_df = rental_data[['RegionName', 'State', 'Metro', 'CountyName', 'SizeRank', 
                             '2019-10', '2019-11', '2019-12']]

house_rent_df.head()
```

```python
# Saving the house rent data into target file.
tgt_file = './data/target/House_Rent_2029.xlsx'
with pd.ExcelWriter(tgt_file) as writer:
    house_rent_df.to_excel(writer, sheet_name='2029_House_Rent_Data')
```

#### Reading and Writing JSON File.

```python
books = [
    {'Name': 'The Spy', 'Author': 'Paulo Coelho', 'Price': 3.23},
    {'Name': 'The Perfect Us', 'Author': 'Durjoy Dutta', 'Price': 5.34},
    {'Name': 'Origin', 'Author': 'Dan Brown', 'Price': 7.33},
    {'Name': 'Sapience', 'Author': 'Yuval Noah Harari', 'Price': 9.23},
    {'Name': 'The Last Mughal', 'Author': 'William Dalrymple', 'Price': 13.43},
]

book_df = pd.DataFrame(books)
book_df
```

```python
# Writing to a JSON File. orient - records.
JSONfile = './data/target/Book_Record.json'
book_df.to_json(JSONfile, orient='records')
```

```python
# Writing to a JSON File. orient - split.
JSONfile = './data/target/Book_Split.json'
book_df.to_json(JSONfile, orient='split')
```

```python
# Writing to a JSON File. orient - index.
JSONfile = './data/target/Book_Index.json'
book_df.to_json(JSONfile, orient='index')
```

```python
# Writing to a JSON File. orient - columns.
JSONfile = './data/target/Book_Columns.json'
book_df.to_json(JSONfile, orient='columns')
```

```python
# Writing to a JSON File. orient - values.
JSONfile = './data/target/Book_Values.json'
book_df.to_json(JSONfile, orient='values')
```

```python
# Writing to a JSON File. orient - table.
JSONfile = './data/target/Book_Table.json'
book_df.to_json(JSONfile, orient='table')
```

```python
# Reading the JSON file data into a DataFrame.
JSONfile = './data/source/test_data.json'
df = pd.read_json(JSONfile)
df
```
