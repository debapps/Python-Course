# Indexing, Slicing of NumPy Array

- **Indexing**: Indexing refers to accessing elements of the array. Index starts from 0.

- **Slicing**: Slicing refers to extracting a portion of existing array. A slice object is of the form **start:end:step**. The **end** element are not included in result of slicing.

```python
import numpy as np
```

### 1-D Array

```python
x = np.arange(10, 50, 3)
print(x)

# Indexing
print(x[0])
print(x[3])
print(x[-1])
print(x[-4])

# Slicing.
print(x[1:6])
print(x[1:6:2])
print(x[::-1]) # Reverse of array.
print(x[3:])
```

#### Slicing Object.

```python
x = np.arange(10, 50, 3)
slice_obj = slice(3, 8)

x_slice = x[slice_obj]
x_slice_cp = x[slice_obj].copy()

print('Original:\n', x, '\nSlice:\n', x_slice, '\nCopy of Slice:\n', x_slice_cp)

print('\nAfter Manipilation...\n')
x[slice_obj] = 0

print('Original:\n', x, '\nSlice:\n', x_slice, '\nCopy of Slice:\n', x_slice_cp)
```

### 2-D Array

```python
data = np.arange(10, 45, 3).reshape(3, 4)
data
```

```python
# Indexing.
print(data[1])
print(data[1, 2])
print(data[-2, -1])

# Slicing.
print(data[1:2, 1:3])
print(data[1:, 1:3])
print(data[1::-1, 1:3])
print(data[:, 2:3])
```

```python
# Slice array manipilation.
data = np.arange(10, 45, 3).reshape(3, 4)
data_slice = data[1:, 2:]
data_slice_cp = data[1:, 2:].copy()
print('Original:\n', data, '\nSlice:\n', data_slice, '\nCopy of Slice:\n', data_slice_cp)

print('\nAfter Slice Manipulation...\n')
# Manipulation of data slice.
data[1:, 2:] = 1
print('Original:\n', data, '\nSlice:\n', data_slice, '\nCopy of Slice:\n', data_slice_cp)
```

### Higher Dimentional Array

```python
info = np.arange(1, 53, 3).reshape(3, 2, 3)
info
```

```python
print(info[1])
print(info[1, :, 1])
print(info[1:, 1, :])
```

#### Iterating using 'nditer'

- **nditer** method of numpy creates an iterator, which enable accessing each element one after the other.

```python
x = np.array([[0,1], [2, 3]])
for a in np.nditer(x):
    print(a)
```

#### Boolean Indexing

- Checking if every element of an array satisfies a condition, results in a Boolean array.

- This Boolean array can be used as index to filter elements that satisfy the condition.

```python
x = np.arange(1,11).reshape(2,5)
print('Original:\n', x)
cond = x % 2 == 0
print('\nBoolean Result:\n', cond)
print('\nBoolean Indexed Array:\n', x[cond])
```

##### Excercise

Write a fuction **array_index** which accepts three numbers **n, n_row, n_col** and performs the array operations given below:

- Create a array **x** of shape **(n_row, n_col)** having first **n** natural numbers.

- Print elements of last row.

- Print elements of middle column.

- Print elements, overlapping first two rows and last three columns.

```python
def array_index(n, n_row, n_col):
    x = np.arange(n).reshape(n_row, n_col)

    print('Array:\n', x)

    print('\nElements of the last row: ', x[-1])

    mid_col = n_col // 2

    print('\nElements of the middle column: ', x[:, mid_col])
    print('\nElements overlapping first two rows and last three columns:\n', x[:2, -3:])

n = 40
n_row = 4
n_col = 10

array_index(n, n_row, n_col)
```
